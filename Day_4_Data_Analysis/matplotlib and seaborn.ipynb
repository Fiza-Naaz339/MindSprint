# matplotlib

```python
import matplotlib.pyplot as plt
print(plt.__package__)
```

```python
# Line graph
plt.figure(figsize=(10,6))
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.plot(x,y,marker='o',linestyle='-',color='b')
plt.xlabel("X-axis")
plt.ylabel("y-lable")
plt.title("Line Plot")
plt.grid()
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10,6))
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.scatter(x,y,marker='*',color='r')
plt.xlabel("X-axis")
plt.ylabel("Y-lable")
plt.title("Line Plot")
plt.grid()
plt.show()
```

```python
# Bar graph
cat = ['A','B','C','D']
val = [5,7,8,9]
plt.bar(cat,val,color="purple")
plt.xlabel("categories")
plt.ylabel("Values")
plt.title("Bar graph")
plt.grid()
plt.show()
```

```python
# Histogram

import numpy as np

data = np.random.randn(1000)
plt.hist(data,bins=30,color='blue',edgecolor='black')
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
# pie chart

size = [20,30,25,25]
plt.pie(size,labels=['js','java','python','C#'],colors=['blue','red','green','orange'],autopct='%1.1f%%')
plt.show()
```

# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    "x":[1,2,3,4,5],
    "y":[10,20,15,25,30]
})

sns.lineplot(x="x",y="y",data=data)
plt.title("Seaborn Line Plot")
plt.show()
```
