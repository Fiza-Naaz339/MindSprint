```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('employee_salary.csv')

# 1. Display Basic Information About the Dataset
print("Basic Information about the Dataset:")
print(df.info())
print("\nFirst 5 rows of the dataset:")
print(df.head())

# 2. Find the Average Salary in Each Department
average_salary = df.groupby('department')['salary'].mean()
print("\nAverage Salary in Each Department:")
print(average_salary)

# 3. Find the Highest and Lowest Salary in the Company
highest_salary = df['salary'].max()
lowest_salary = df['salary'].min()
print(f"\nHighest Salary in the Company: {highest_salary}")
print(f"Lowest Salary in the Company: {lowest_salary}")

# 4. Find the Employee with the Maximum Salary
max_salary_employee = df.loc[df['salary'].idxmax()]
print("\nEmployee with the Maximum Salary:")
print(max_salary_employee)

# 5. Count the Number of Employees in Each Department
employee_count = df['department'].value_counts()
print("\nNumber of Employees in Each Department:")
print(employee_count)

# 6. Plot a Bar Chart Showing the Number of Employees in Each Department
plt.figure(figsize=(10, 6))
employee_count.plot(kind='bar', color='skyblue')
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()

# 7. Plot a Pie Chart Showing Salary Distribution by Department
salary_distribution = df.groupby('department')['salary'].sum()
plt.figure(figsize=(8, 8))
salary_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Salary Distribution by Department')
plt.ylabel('')  # Hide the y-label
plt.show()

# 8. Scatter Plot Showing Salary vs Experience
plt.figure(figsize=(10, 6))
plt.scatter(df['experience_year'], df['salary'], alpha=0.5, color='orange')
plt.title('Salary vs Experience')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.grid()
plt.show()

# 9. Find Employees Who Earn More Than the Average Salary
average_salary_value = df['salary'].mean()
high_earners = df[df['salary'] > average_salary_value]
print("\nEmployees Who Earn More Than the Average Salary:")
print(high_earners)

# 10. Sort Employees Based on Salary in Descending Order
sorted_employees = df.sort_values(by='salary', ascending=False)
print("\nEmployees Sorted by Salary in Descending Order:")
print(sorted_employees)
```
