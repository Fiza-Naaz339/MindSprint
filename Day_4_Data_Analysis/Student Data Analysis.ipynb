```python
df['Total'] = df[['Math','Science','English']].sum(axis=1)
```

```python
df['Total']
```

```python
# To verify Total
df[['Name','Math','Science','English','Total']].head()
```

```python
def assign_garde(total):
    if total>=250:
        return 'A'
    elif total>=200:
        return 'B'
    elif total>=150:
        return 'C'
    else:
        return 'F'
```

```python
df['Grade'] = df['Total'].apply(assign_garde)
print(df[['Name','Total','Grade']])
```
